```python
def noise(img,snr):
    h=img.shape[0]
    w=img.shape[1]
    img1=img.copy()
    sp=h*w
    NP=int(sp*(1-snr))
    for i in range (NP):
        randx=np.random.randint(1,h-1)
        randy=np.random.randint(1,w-1)
        
        if np.random.random()<=0.5:
            img1[randx,randy]=0
        else:
            img1[randx,randy]=255
    return img1

img = cv.imread("/kaggle/input/rakauas/Download-Free-Assassins-Creed-Black-Flag-Wallpaper-1122x631.jpg",1)

img_out = noise(img, 0.6)
plt.figure(figsize=(20, 20))
plt.subplot(1,2,1)
plt.imshow(img,cmap="gray")
plt.subplot(1,2,2)
plt.imshow(img_out,cmap="gray")
plt.show()
```

```python
img = cv2.imread("/kaggle/input/rakauas/Download-Free-Assassins-Creed-Black-Flag-Wallpaper-1122x631.jpg",1)
img_noised = noise(img, 0.6)
median = cv2.medianBlur(img_noised, 5)

plt.figure(figsize=(20, 20))
plt.subplot(1,2,1)
plt.imshow(img_noised,cmap="gray")
plt.subplot(1,2,2)
plt.imshow(median,cmap="gray")
plt.show()
```
